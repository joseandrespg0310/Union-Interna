# Union de DATAFRAMES

```python
dest = pd.read_csv("E:/Descargas/dest.csv")
dest.head()
```

```python
#Carga DataFrame tips.csv
tips = pd.read_csv("E:/Descargas/tips.csv")
tips.head()
```

# Creamos una Union Interna

```python
#Despliega numero de empleado destino y cantidad usando Inner Join
df_inner=pd.merge(dest,tips, on='EmpNr',how='inner')
df_inner.head()
```

# Union Externa Completa

```python
# Une data frames usando union externa completa
df_outer = pd.merge(dest,tips,on='EmpNr',how='outer')
df_outer.head()
```

# Union externa derecha

```python
# Une dataframes usando union externa dereha
df_right = pd.merge(dest,tips, on='EmpNr',how='right')
df_right.head()
```

```python
# Hacer union izquiersa
df_l = pd.merge(dest,tips, on='EmpNr',how='left')
df_l .head()
```
